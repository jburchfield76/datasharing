```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels.api as sm

# Simulated data
np.random.seed(42)
ad_spend = np.random.randint(100, 1000, 100)  # Ad spend in dollars
purchases = np.random.poisson(lam=(ad_spend / 200), size=100)  # More ad spend, more purchases

# Model (Using statsmodels for Poisson regression)
X = sm.add_constant(ad_spend)  # Add intercept
poisson_model = sm.GLM(purchases, X, family=sm.families.Poisson(link=sm.families.links.log()))
poisson_results = poisson_model.fit()
y_pred = poisson_results.predict(X)

# Plot results
plt.scatter(ad_spend, purchases, label="Actual Purchases")
plt.scatter(ad_spend, y_pred, color='red', label="Predicted Purchases")
plt.xlabel("Ad Spend ($)")
plt.ylabel("Number of Purchases per Month")
plt.legend()
plt.title("Poisson Regression (Log Link)")
plt.show()
```
