```python
%%R
# Create sample data
set.seed(42)
x <- rnorm(50)
y <- 3 * x + rnorm(50)

# Fit a linear model
model <- lm(y ~ x)
summary(model)

# Plot with regression line
plot(x, y, main = "Linear Regression", col = "blue")
abline(model, col = "red")
```
