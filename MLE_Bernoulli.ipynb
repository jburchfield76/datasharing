```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Generate synthetic data from a Bernoulli distribution
np.random.seed(42)  # For reproducibility
true_p = 0.7        # True probability of success
n = 100             # Number of trials

data = np.random.binomial(n=1, p=true_p, size=n)  # Generate Bernoulli trials (0 or 1)

# Step 2: Compute MLE for p
mle_p = np.mean(data)  # MLE estimate for p (sample mean)

# Print results
print(f"True p: {true_p}, MLE estimated p: {mle_p:.4f}")

# Step 3: Visualizing the results
x = [0, 1]  # Possible outcomes in a Bernoulli distribution
true_pmf = [1 - true_p, true_p]  # True PMF: P(X=0) and P(X=1)
mle_pmf = [1 - mle_p, mle_p]  # MLE estimated PMF

plt.bar(x, true_pmf, alpha=0.5, color='r', label="True Distribution", width=0.4)
plt.bar(np.array(x) + 0.4, mle_pmf, alpha=0.5, color='b', label="MLE Estimated Distribution", width=0.4)
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("MLE Estimation of Bernoulli Distribution")
plt.legend()
plt.show()
```
