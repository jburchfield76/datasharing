```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Simulated data
np.random.seed(42)
income = np.random.randint(20000, 100000, 100)  # Income in dollars
purchase = np.random.choice([0, 1], size=100, p=[0.6, 0.4])  # 40% buy, 60% don't

# Model
X = income.reshape(-1, 1)
y = purchase
log_model = LogisticRegression()
log_model.fit(X, y)
y_pred_prob = log_model.predict_proba(X)[:, 1]  # Predicted probabilities

# Plot results
plt.scatter(income, purchase, label="Actual Purchases", alpha=0.5)
plt.scatter(income, y_pred_prob, color='red', label="Predicted Probabilities", alpha=0.5)
plt.xlabel("Income ($)")
plt.ylabel("Probability of Purchase")
plt.legend()
plt.title("Logistic Regression (Logit Link)")
plt.show()
```
