```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Generate synthetic data
np.random.seed(42)  # Ensure reproducibility
true_mu = 50        # True mean
true_sigma = 10     # True standard deviation
n = 100             # Sample size

data = np.random.normal(loc=true_mu, scale=true_sigma, size=n)  # Generate data

# Step 2: Compute MLE estimates
mle_mu = np.mean(data)  # MLE estimate for mean
mle_sigma_sq = np.mean((data - mle_mu) ** 2)  # MLE estimate for variance (not n-1 correction)
mle_sigma = np.sqrt(mle_sigma_sq)  # Standard deviation estimate

# Print results
print(f"True mean (μ): {true_mu}, MLE estimated mean: {mle_mu:.2f}")
print(f"True std dev (σ): {true_sigma}, MLE estimated std dev: {mle_sigma:.2f}")

# Step 3: Visualizing the results
x = np.linspace(min(data)-10, max(data)+10, 1000)
true_pdf = stats.norm.pdf(x, loc=true_mu, scale=true_sigma)  # True distribution
mle_pdf = stats.norm.pdf(x, loc=mle_mu, scale=mle_sigma)  # Estimated distribution

plt.hist(data, bins=15, density=True, alpha=0.6, color='g', label="Sample Data")
plt.plot(x, true_pdf, 'r--', label="True Distribution")
plt.plot(x, mle_pdf, 'b-', label="MLE Estimated Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("MLE Estimation of Normal Distribution")
plt.legend()
plt.show()
```
