```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Generate synthetic data from an exponential distribution
np.random.seed(42)  # For reproducibility
true_lambda = 0.5   # True rate parameter (Mean = 1/lambda = 2)
n = 100             # Number of samples

data = np.random.exponential(scale=1/true_lambda, size=n)  # Generate data

# Step 2: Compute MLE for lambda
mle_lambda = 1 / np.mean(data)  # MLE estimate for lambda

# Print results
print(f"True lambda: {true_lambda}, MLE estimated lambda: {mle_lambda:.4f}")
print(f"True mean (1/lambda): {1/true_lambda}, Sample mean: {np.mean(data):.4f}")

# Step 3: Visualizing the results
x = np.linspace(0, max(data), 1000)
true_pdf = stats.expon.pdf(x, scale=1/true_lambda)  # True exponential distribution
mle_pdf = stats.expon.pdf(x, scale=1/mle_lambda)  # MLE estimated exponential distribution

plt.hist(data, bins=20, density=True, alpha=0.6, color='g', label="Sample Data")
plt.plot(x, true_pdf, 'r--', label="True Distribution")
plt.plot(x, mle_pdf, 'b-', label="MLE Estimated Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("MLE Estimation of Exponential Distribution")
plt.legend()
plt.show()
```
