```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Simulated data
np.random.seed(42)
square_feet = np.random.randint(800, 4000, 100)  # House size in square feet
price = 50000 + 100 * square_feet + np.random.normal(0, 25000, 100)  # House price with some noise

# Model
X = square_feet.reshape(-1, 1)  # Reshape for sklearn
y = price
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot results
plt.scatter(square_feet, price, label="Actual Prices")
plt.plot(square_feet, y_pred, color='red', label="Predicted Prices")
plt.xlabel("Square Feet")
plt.ylabel("Price ($)")
plt.legend()
plt.title("Linear Regression (Identity Link)")
plt.show()
```
