```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Simulated data
np.random.seed(42)
square_feet = np.random.randint(800, 4000, 100)  # House size in square feet, sample of n=100, between 800-4000 sq ft
price = 50000 + 100 * square_feet + np.random.normal(0, 44000, 100)  # House price with some noise

# Model
X = square_feet.reshape(-1, 1)  # Reshape for sklearn
y = price
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
from sklearn.metrics import r2_score, mean_squared_error

# Calculate R-squared and MSE
r_squared = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

# Display results
print(f"R-squared: {r_squared:.4f}")
print(f"Mean Squared Error: {mse:,.2f}")
print(f"Root Mean Squared Error: {rmse:,.2f}")


# Plot results
plt.scatter(square_feet, price, label="Actual Prices")
plt.plot(square_feet, y_pred, color='green', label="Pred. Prices")
plt.xlabel("Square Feet")
plt.ylabel("Price ($)")
plt.legend()
plt.title("Linear Regression (Identity Link)")
plt.show()
```
