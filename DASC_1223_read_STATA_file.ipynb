```python
from urllib import request
zip_url = 'https://gss.norc.org/content/dam/gss/get-the-data/documents/stata/2024_stata.zip'
request.urlretrieve(zip_url, filename='gss_stata_with_codebook.zip') #go to site, rt-click most recent year, copy-paste link to file

from zipfile import ZipFile
with ZipFile('gss_stata_with_codebook.zip', mode='r') as zip:
    zip.extractall()
    for file in zip.infolist():
        print(file.filename)

import pyreadstat
gss_stata_filename = '2024/GSS2024.dta'    #see file listed below, whatever year you used

```

```python
df, meta = pyreadstat.read_dta(gss_stata_filename) #get the metadata container
```

```python
print(df.attrs)  # Usually empty unless you added attributes manually
```

```python
#print the attributes from the data frame
# Print column (variable) names
print(meta.column_names)

# Print column labels (descriptions from Stata)
print(meta.column_labels)  # Dictionary: column_name → label

# Print value labels (for categorical variables)
print(meta.value_labels)   # Dictionary: label_set_name → {value: label}
```

```python
#build a dataframe from the metadata

meta_df = pd.DataFrame({
    'variable_name': meta.column_names,
    'variable_label': meta.column_labels,
})

meta_df.head()  # View the first few rows
```
